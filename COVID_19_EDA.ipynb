## Plots & Charts

### Active cases according to date in India

```python
x=India['date']
y=India['active_cases']

plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.xlabel('Date')
plt.ylabel('Active Cases')
plt.title('Active Cases According to Date')
```

### Comparison of Total Deaths in India VS USA

```python
India=df[df['country']=='India']
USA=df[df['country']=='USA']

x=India['daily_new_deaths'].sum()
y=USA['daily_new_deaths'].sum()

plt.bar(['India','USA'],[x,y],color=['orange','blue'])
plt.xlabel("Countries")
plt.ylabel("Total no. of Cases")
plt.title(' Total Deaths India VS USA')
plt.show()
```

### Top 10 Countries by total COVID-19 Cases

```python
data=df.groupby('country')['daily_new_cases'].sum()
```

```python
top10 = data.sort_values(ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(top10.index, top10.values, color="crimson")
plt.title('Top 10 countries by total COVID-19 Cases')
plt.xlabel('Countries')
plt.ylabel('Total Cases')
plt.show()
```

### Top 10 countries by total deaths

```python
country_deaths = df.groupby('country')['cumulative_total_deaths'].max().sort_values(ascending=False)

plt.figure(figsize=(10,6))
country_deaths.head(10).plot(kind='bar', color='salmon')
plt.title('Top 10 Countries by Total Deaths', fontsize=14)
plt.ylabel('Total Deaths')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
    
```

### Line plots of total cases for multiple countries.

```python
df['date'] = pd.to_datetime(df['date'])
countries = ["India", "United States", "Brazil", "Russia", "United Kingdom"]
plt.figure(figsize=(12,6))


for country in countries:
    country_data = df[df['country'] == country]
    plt.plot(country_data['date'], 
             country_data['cumulative_total_cases'], 
             label=country)

plt.title("Cumulative Total COVID-19 Cases - Country Comparison", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Cumulative Total Cases", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
